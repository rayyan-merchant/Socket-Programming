```python
import socket
import threading
```

```python
def receive():
    while True:
        try:
            msg = s.recv(1024).decode()
            print(msg, flush=True)
        except:
            break
```

```python
s = socket.socket()
s.connect(('localhost', 9999))

name = input("Enter name: ")
s.send(name.encode())

threading.Thread(target=receive, daemon=True).start()
```

```python
while True:
    msg = input()

    if msg.lower() == "exit":
        s.send("exit".encode())
        print("Disconnected")
        s.close()
        break

    s.send(f"{name}: {msg}".encode())
```
