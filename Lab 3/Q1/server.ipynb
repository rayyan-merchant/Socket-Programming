```python
import socket 
```

```python
s = socket.socket()
print('socket created')
```

```python
s.bind(('localhost', 9999))

s.listen(1)
print("Server waiting for connection...")
```

```python
c, addr = s.accept()
print("got connectedd from", addr)
```

```python
filename = c.recv(1024).decode()

filesize = int(c.recv(1024).decode())

print(f"Receiving {filename} ({filesize} bytes)")
```

```python
received = 0

with open("received " + filename, "wb") as f:
    while received < filesize:
        data = c.recv(1024)
        if not data:
            break
        f.write(data)
        received += len(data)

print("File received successfully.")
```

```python
c.close()
s.close()
```
