```python
import socket
import threading
import os
```

```python
def receive():
    while True:
        try:
            msg = s.recv(1024).decode()

            if msg.startswith("FILE:"):
                parts = msg.split(":")
                filename = parts[1]
                filesize = int(parts[2])

                f = open("received_" + filename, "wb")
                remaining = filesize

                while remaining > 0:
                    data = s.recv(1024)
                    f.write(data)
                    remaining = remaining - len(data)

                f.close()
                print("Received file:", filename)

            else:
                print(msg)

        except:
            break
```

```python
s = socket.socket()
s.connect(('localhost', 9999))
```

```python
name = input("Enter name: ")
s.send(name.encode())
```

```python
t = threading.Thread(target=receive)
t.daemon = True
t.start()
```

```python
while True:
    msg = input()

    if msg.lower() == "exit":
        s.send("exit".encode())
        print("Disconnected")
        s.close()
        break

    if msg.startswith("sendfile "):
        parts = msg.split(" ")
        filepath = parts[1]

        if not os.path.exists(filepath):
            print("File does not exist")
            continue

        filename = os.path.basename(filepath)
        filesize = os.path.getsize(filepath)

        s.send(("FILE:" + filename + ":" + str(filesize)).encode())

        f = open(filepath, "rb")
        while True:
            data = f.read(1024)
            if not data:
                break
            s.send(data)
        f.close()

    else:
        s.send((name + ": " + msg).encode())
```
